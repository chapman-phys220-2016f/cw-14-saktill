From the data file we get all of the data reported from Manzanita Lake.

```python
mLake = df[df['STATION_NAME']=='MANZANITA LAKE CA US']
mLake
```

Below we use a set of confusing looking boolean tests on the data of the dates and maximum and minimum temperatures.
We then print out the day that these respective temperatures occur on in the tuple format (date, temperature).

```python
tMax = (int(mLake[mLake['TOBS'] == mLake['TOBS'].max()]['DATE']), int(mLake[mLake['TOBS'] == mLake['TOBS'].max()]['TOBS']))
tMin = (int(mLake[mLake['TOBS'] == mLake['TOBS'].min()]['DATE']), int(mLake[mLake['TOBS'] == mLake['TOBS'].min()]['TOBS']))
```

```python
tMax
```

```python
tMin
```

```python
plt.bar([0,1],[tMin[1],tMax[1]],tick_label=['tMin','tMax'])
plt.ylabel('Degrees Celsius E-1')
plt.show()
```

We chose Mount Wilson because we thought it would be interesting to look at a mountainous area for this precipitation data.<br\>
Below we're getting the dates that are listed inbetween 06/2015 and 07/2015 for Mount Wilson and then plotting the results of this. 

```python
mtWil = df[df['STATION_NAME']=='MOUNT WILSON CBS CA US']
d1 = mtWil[mtWil['DATE']>20150600]
d2 = d1[d1['DATE']<20150700]
plt.plot(np.array(d2['DATE'])-20150600,np.array(d2['PRCP']))
plt.xlabel('June 2015')
plt.ylabel('Rainfall (mm E-1)')
plt.show()
```
